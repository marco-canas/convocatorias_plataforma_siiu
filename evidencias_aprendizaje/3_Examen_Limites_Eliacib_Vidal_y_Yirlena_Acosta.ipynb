Un fabricante puede producir grabadoras a un costo de $US 40$ la unidad. Se estima que si las  grabadores se venden a un precio de $ùë•$ d√≥lares cada una, los consumidores comprar√°n   $120 ‚àí ùë•$ de estas al mes.

 Exprese el beneficio de la compa√±√≠a como una funci√≥n del precio unitario, dibuje la gr√°fica de esta funci√≥n y util√≠cela para estimar el precio √≥ptimo de venta, el beneficio √≥ptimo, y determine el dominio y recorrido de esta funci√≥n.



Ejercicio 16.20

Planteamiento del problema:


  El costo de producci√≥n de una grabadora es de \$40 por unidad.
 La cantidad demandada depende del precio de venta. Se estima que los consumidores comprar√°n $120 - x$ unidades al mes si el precio de venta es $x$ d√≥lares cada una.

 Se utiliza la ecuacion cuadratica:


$$F(x)= ax^2+bx+c$$

Expresi√≥n del beneficio de la compa√±√≠a como funci√≥n del precio unitario:


\begin{align*}
    &\text{Ingresos} = x(120 - x) \\
    &\text{Costos} = 40(120 - x) \\
    &\text{Beneficio} = \text{Ingresos} - \text{Costos} = x(120 - x) - 40(120 - x)
\end{align*}

Simplificaci√≥n de la funci√≥n de beneficio:

\begin{align*}
    \text{Beneficio} &= 120x - x^2 - 4800 + 40x \\
    &= 160x - x^2 - 4800
\end{align*}

Identificaci√≥n del v√©rtice de la par√°bola (punto m√°ximo de la funci√≥n):

\begin{align*}
    &x = -\frac{b}{2a}, \quad \text{donde } a = -1 \text{ y } b = 160 \\
    &x = -\frac{160}{2(-1)} = 80 \\
    &y = 160(80) - (80)^2 - 4800 = 12800 - 6400 - 4800 = 1600
\end{align*}

Por lo tanto, el v√©rtice de la par√°bola es $(80, 1600)$, lo que indica que el precio √≥ptimo de venta es de \$80 y el beneficio √≥ptimo es de \$1600.

Determinaci√≥n de los puntos de intersecci√≥n con el eje $x$ (dominio):

\begin{align*}
    0 &= 160x - x^2 - 4800 \\
    &= (x - 40)(120 - x)
\end{align*}

Esto nos da dos soluciones: $x = 40$ y $x = 120$. Sin embargo, el valor de $x$ no puede ser mayor que 120 (ya que no tendr√≠a sentido para la demanda negativa), entonces solo consideramos $x = 40$.

Conclusi√≥n:


  El precio √≥ptimo de venta para maximizar el beneficio es \$80.
  El beneficio m√°ximo es de \$1600.
  El dominio de la funci√≥n es $x \in [12, 40]$ y el recorrido es $y \in [-4800, 1600]$.



```python
import numpy as np
import plotly.graph_objects as go

# Definir la funci√≥n de beneficio
def beneficio(x):
    return 160*x - x**2 - 4800

# Calcular el v√©rtice de la par√°bola (m√°ximo beneficio)
x_max = 80
y_max = beneficio(x_max)

# Calcular los puntos de inicio y declive
x_inicio = 40
y_inicio = beneficio(x_inicio)
x_declive = 120
y_declive = beneficio(x_declive)

# Generar valores de x para la gr√°fica
x = np.linspace(10, 130, 400)
# Calcular los valores correspondientes de y
y = beneficio(x)

# Ajustar los l√≠mites del eje y para que los valores positivos sean m√°s visibles
y_min = min(y)
y_max = max(y)
y_range = y_max - y_min
y_lim = (y_min - 0.1 * y_range, y_max + 0.1 * y_range)

# Crear la figura interactiva
fig = go.Figure()

# Agregar la l√≠nea de beneficio
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Beneficio'))

# Agregar los puntos de inter√©s
fig.add_trace(go.Scatter(x=[x_max, x_inicio, x_declive],
                         y=[y_max, y_inicio, y_declive],
                         mode='markers',
                         marker=dict(color=['red', 'green', 'orange'], size=10),
                         text=['M√°ximo (80, 1600)', 'Inicio (40, 3200)', 'Declive (120, -1200)'],
                         hoverinfo='text'))

# Establecer el dise√±o del gr√°fico
fig.update_layout(
    title='Gr√°fico del beneficio en funci√≥n del precio de venta',
    xaxis=dict(title='Precio de venta ($)'),
    yaxis=dict(title='Beneficio ($)', range=y_lim),  # Ajustar los l√≠mites del eje y
    hovermode='closest'
)

# Mostrar la gr√°fica interactiva
fig.show()
```
