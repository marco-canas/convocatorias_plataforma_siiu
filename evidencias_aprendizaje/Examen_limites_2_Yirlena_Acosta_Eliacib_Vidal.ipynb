gr√°fica de esta funci√≥n y util√≠cela para estimar el precio √≥ptimo de venta, el beneficio √≥ptimo, y determine el dominio y recorrido de esta funci√≥n.

Hallar:
la expresion del costo, beneficio e ingreso

$ I= x(10000-x) $  
$ I= 10000x-x^2$  

$ C= 1200 (10000-x)$  
$ C= 12000000-1200x$


$ B(x)=(10000x-x^2)-(12000000-1200x)$  
$ B(x)= x^2+10000x+1200x-12000000$     
      $ = x^2+11200x-12000000$   

se toma en cuenta la expresion cuadratica:
$$ f(x)= ax^2+bx-c $$   








Identificamos el vertice de la paravola

$11200/2=5600$  


hacemos uso de la funcion cuadratica:
$ x^2-11200x-12000000$   

$11200(5600)-(5600)^2-12000000$  
$ =62720000-31360000-12000000= 19360000$

siendo asi el vertice de la paravola:   

$[5600,19360000]$  

$ x(dominio)= 11200-x^2-12000000 $      
$            =(x-1200),(10000-x) $

```python
import plotly.graph_objs as go

# Datos proporcionados
precios = [1200,1600,2000,2400,2800,3200,3600,4000,4400,4800,5200,5600,6000,6400,6800,7200,7600,8000,8400,8800,9200,9600,10000]
beneficios_millones = [0,3360000,6400000,9120000,11520000,13600000,15360000,16800000,17920000,18720000,19200000,19360000,19200000,18720000,17920000,16800000,15360000,13600000,11520000,9120000,6400000,3360000,0]

# Identificamos el punto m√°s alto
max_beneficio = max(beneficios_millones)
indice_maximo = beneficios_millones.index(max_beneficio)
precio_maximo = precios[indice_maximo]

# Creamos la gr√°fica interactiva
fig = go.Figure()

# A√±adimos la l√≠nea principal
fig.add_trace(go.Scatter(x=precios, y=beneficios_millones, mode='lines+markers', name='Beneficio',
                         hoverinfo='x+y', line=dict(color='violet')))

# A√±adimos el punto m√°s alto con un color diferente
fig.add_trace(go.Scatter(x=[precio_maximo], y=[max_beneficio], mode='markers', name='M√°ximo',
                         marker=dict(color='purple', size=10), hoverinfo='x+y'))

# T√≠tulos y etiquetas
fig.update_layout(title='Gr√°fica Interactiva de Precios vs Beneficio',
                  xaxis_title='Precio (COP)',
                  yaxis_title='Beneficio (millones)',
                  yaxis=dict(tickformat=',d'))

# Mostrar la gr√°fica
fig.show()
```

Utilizamos la grafica para estimar el precio √≥ptimo de venta y el beneficio √≥ptimo:

$$P= [PMin: 1200, PMax: 5200] $$ entendiento que el precio sugerido para el maximo beneficio es de 5200.

$$B= [BMin: 3360000, BMax: 19360000] $$ se entiende as√≠ que el maximo beneficio es de 19360000.  




#CONCLUSIONES:
- Punto de equilibrio: Observamos que el punto donde la par√°bola alcanza su punto m√°ximo (representando la m√°xima ganancia) ocurre alrededor de x = 5600.
- Optimizaci√≥n de costos y beneficios: La forma de la par√°bola indica que, al principio, a medida que aumenta la cantidad producida (x), el beneficio aumenta a un ritmo creciente. Esto sugiere que la empresa debe equilibrar el aumento de la producci√≥n con la optimizaci√≥n de los costos y beneficios.
- An√°lisis de sensibilidad: La gr√°fica tambi√©n muestra que el beneficio no es lineal con respecto a la cantidad producida. Esto implica que peque√±os cambios en la cantidad producida pueden tener un impacto significativo en el beneficio total. Por lo tanto, la empresa debe realizar un an√°lisis de sensibilidad para comprender c√≥mo los cambios en la cantidad producida pueden afectar su rentabilidad.
- Importancia de la precisi√≥n en la toma de decisiones: Dado que la cantidad √≥ptima para alcanzar la m√°xima ganancia por unidad es de 5600 bloques, la empresa debe esforzarse por tomar decisiones precisas en cuanto a la cantidad de producci√≥n.
